### Loading Another Dataset to perform Various operations

```python
data=pd.read_csv("C:/Users/Aniket/Downloads/TravelChoices.csv")
data.head()
```

```python
df=df.fillna(0)
df
```

```python
data=data.fillna(0)
data
```

```python
df['Age'] = df['Age'].astype(int)  # Convert to int
df
```

```python
merge_data = df.merge(data, on="Age")
merge_data.head()
```

```python
group_data = merge_data.groupby("Age")["Education"].count()
group_data.head()
```

```python
merge_data = merge_data.drop_duplicates(subset=['Age','City'])
merge_data
```

```python
Pivot=merge_data.pivot_table(index="PassengerId",columns="Ticket",values="Fare")
Pivot
```

```python
Pivot=Pivot.fillna(0)
Pivot
```
